```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv("ToyotaCorolla.csv", encoding='latin1')

# Display the first few rows of the dataset
print(data.head())

# Check for missing values
print(data.isnull().sum())
```

```python
import matplotlib.pyplot as plt

# Histogram of car prices
plt.figure(figsize=(10, 6))
plt.hist(data['Price'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Car Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Scatter plot of car age vs. mileage
plt.figure(figsize=(10, 6))
plt.scatter(data['Age_08_04'], data['KM'], color='green', alpha=0.5)
plt.title('Scatter Plot of Car Age vs. Mileage')
plt.xlabel('Age')
plt.ylabel('Mileage')
plt.grid(True)
plt.show()
```
